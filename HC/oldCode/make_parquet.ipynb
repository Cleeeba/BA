```python
from IPython.display import display
import pandas as pd
import os
import csv

```

```python
def read_in(string):
    df = pd.read_csv(string, compression='gzip', header=0, sep='\t', quotechar='"', on_bad_lines='skip', quoting=csv.QUOTE_NONE)
    first, *middle, last = string.split("_")
    df.to_parquet( last +'.parquet')
    print(pd.read_parquet(last +'.parquet'))
```

```python
directory = 'C:/Users/bincl/BA-Thesis/Dataset/test2Gram'

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        read_in(f) 
```
